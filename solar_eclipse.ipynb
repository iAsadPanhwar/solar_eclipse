```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\asada\\Desktop\\Projects\\solar_elipse\\solar.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(columns = ['Duration in Seconds'], inplace = True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
na_col = ['Central Duration Seconds', 'Normalized Duration','Normalized Path Width', 'EII', 'ESC Wide-Scale Moving Average', 'ESC Moving Average']
```

```python
df[na_col] = df[na_col].fillna(df[na_col].median())
```

```python
df.isna().sum()
```

```python
df['Path Width (km)'] = pd.to_numeric(df['Path Width (km)'], errors='coerce')
df['Central Duration'] = pd.to_numeric(df['Central Duration'], errors='coerce')
```

```python
df.isna().sum()
```

```python
df['Path Width (km)'].fillna(-1, inplace=True)
df['Central Duration'].fillna(-1, inplace=True)
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['Eclipse Longitude'], y=df['Eclipse Latitude'], hue=df['Eclipse Type'], style=df['Eclipse Type'], s=50)
plt.title('Geographic Distribution of Eclipses')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
eclipse_counts = df['Year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
eclipse_counts.plot(kind='line')
plt.title('Frequency of Eclipses Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Eclipses')
plt.grid(True)
plt.show()
```
